### Dependencies/Imports

```python
#Dependencies/Imports
import os
import csv
import time
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

### Lists to store data

```python
# Lists to store data
compound_score=[]
pos_score=[]
neu_score=[]
neg_score=[]
```

### Open Datafile and Process Rows

```python
# Set and Open Path to datafile
lyrics_csv = os.path.join("data_output.csv")
df = pd.read_csv(lyrics_csv, encoding="mbcs",index_col=0)


#Process rows
for index, target_string in enumerate(df["Lyrics"]):  
    #print(item)
    # Target String Setting
    #target_string = row[5]
    #print(target_string)
    try:
        # Run analysis
        print("Trying ID: " + str(index))
        compound = analyzer.polarity_scores(target_string)["compound"]
        pos = analyzer.polarity_scores(target_string)["pos"]
        neu = analyzer.polarity_scores(target_string)["neu"]
        neg = analyzer.polarity_scores(target_string)["neg"]

        #if target_string !="" and compound != 0 and neu != 1: #set conditionals to avoid bad rows                   


        # Run analysis from content or song lyrics
        compound_score.append(compound)
        pos_score.append(pos)
        neu_score.append(neu)
        neg_score.append(neg)     
    except:
        print("Retrying ID: " + str(index))
        compound_score.append("")
        pos_score.append("")
        neu_score.append("")
        neg_score.append("")

print(df.head())
```

### Prepare and Send Output to File

```python
# Zip lists together
cleaned_csv = zip(df["Title"],df["Artist"],df["Year"],df["Rank"],df["Lyrics"],df["Source"],df["CheckURL"], compound_score, pos_score, neu_score, neg_score)
#print(cleaned_csv)
# Set variable for output file
output_file = os.path.join("billboard_sentanalys.csv")

#  Open the output file
with open(output_file, "w+", newline="", encoding="UTF-8") as datafile:
    writer = csv.writer(datafile)

    # Write the header row
    writer.writerow(["Title", "Artist", "Year", "Rank","Lyrics","Source","CheckURL","comp_score",
                     "pos_score", "neu_score", "neg_score"])

    # Write in zipped rows
    writer.writerows(cleaned_csv)
```
