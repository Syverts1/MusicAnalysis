```python
import sys
import numpy as np
import random
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import json
import time
import sys
from pprint import pprint
import os
import csv
import seaborn as sns
```

```python
#Songs and Artists with Spotify Consolidatd Data - "artistfinal_with_avgcomp.csv"
```

```python
csv_path = "artistfinal_with_avgcomp.csv"
songs_df=pd.read_csv(csv_path,encoding='ANSI')
songs_df.head()
#len(songs_df)
```

```python
#Popularity vs. Sentiment Plot
plt.figure(figsize=(10,7))
plt.scatter(songs_df["popularity"],songs_df["avgcompscore"])
#plt.scatter(pd.to_numeric(songs_df["popularity"]), pd.to_numeric(songs_df["avgcompscore"]))
sns.set()
plt.grid(True)
plt.title(f"Song Sentiment vs. Popularity (Spotify)", fontsize=20)
plt.axhline(y=0, color = 'black') #adds a horizontal line at zero
plt.ylabel("Average Coumpound Score", fontsize=14)
plt.xlabel("Popularity (Ranking)", fontsize=14)
plt.ylim(-1, 1)
plt.xlim(0, 100)
plt.show()
```

```python
df = songs_df.groupby("popularity")["avgcompscore"].mean().reset_index()
df.head()
#len(df)
```

```python
pop_gt0 = df["popularity"] > 0
df = df[pop_gt0]
df.head()
```

```python
#Popularity vs. Sentiment Plot
x=df["popularity"]
y=df["avgcompscore"]

plt.figure(figsize=(10,7))
plt.scatter(x,y)
#plt.scatter(pd.to_numeric(songs_df["popularity"]), pd.to_numeric(songs_df["avgcompscore"]))
sns.set()
plt.grid(True)
plt.title(f"Song Sentiment vs. Popularity (Spotify)", fontsize=20)
plt.axhline(y=0, color = 'black') #adds a horizontal line at zero
plt.ylabel("Sentiment Average Compound Score", fontsize=14)
plt.xlabel("Popularity (Ranking)", fontsize=14)
#plt.ylim(-1, 1)
#plt.xlim(0, 100)
plt.show()
```

```python
# Billboard Charts Consolidated Data - "billboard_sentanalys.csv"
```

```python
csv_path = "billboard_sentanalys.csv"
songs_df2=pd.read_csv(csv_path,encoding='ANSI')
#songs_df2.head()
```

```python
df = songs_df2.groupby("Rank")["comp_score"].mean().reset_index()
df.head()
#len(df)
```

```python
x=df["Rank"]
y=df["comp_score"]

plt.figure(figsize=(10,7))
sns.set()
plt.grid(True)
plt.scatter(x,y)
plt.title("Song Sentiment vs. Popularity (Billboard Charts)", fontsize=20)
plt.xlabel("Popularity (Ranking) ",fontsize=14)
#plt.axhline(y=0, color = 'black') #adds a horizontal line at zero
plt.ylabel("Sentiment Average Compound Score",fontsize=14)
#plt.xlim(1,100)
#plt.ylim(-1,1)
plt.show()
```

```python
df = songs_df2.groupby("Year")["comp_score"].mean().reset_index()
df.head()
```

```python
x=df["Year"]
y=df["comp_score"]

plt.figure(figsize=(10,7))
sns.set()
plt.grid(True)
plt.scatter(x,y)
plt.title("Song Sentiment Over Time (Release Year - Billboard Charts)", fontsize=20)
plt.xlabel("Release Year", fontsize=14)
plt.ylabel("Sentiment Average Compound Score",fontsize=14)
#plt.xlim(1965,2017)
#plt.ylim(-.5,1)
plt.show()
```

```python
df = songs_df2.groupby("Artist")["Rank"].count().reset_index()
df = df.sort_values(by=["Rank"], axis=0, ascending=False)
#df.head(50)
```

```python
rank_gt11 = df["Rank"] > 11
df = df[rank_gt11]
df = df.sort_values(by=["Rank"], axis=0, ascending=False)
df.head()
```

```python
x=df["Artist"]
y=df["Rank"]

plt.figure(figsize=(10,7))
sns.set()
plt.grid(True)
plt.barh(x,y)
plt.
plt.title("Artist Longevity in Boillboard Charts ", fontsize =20)
plt.xlabel("Popularity (Ranking) ", fontsize=14)
plt.ylabel("Artist", fontsize=14)
#plt.xlim(1,100)
#plt.ylim(-1,1)
plt.show()
```
