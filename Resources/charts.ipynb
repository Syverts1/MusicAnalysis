```python
import sys
import numpy as np
import random
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import json
import time
import sys
from pprint import pprint
import os
import csv
import seaborn as sns
```

```python
csv_path = "artistfinal_with_avgcomp.csv"
songs_df=pd.read_csv(csv_path,encoding='ANSI')
songs_df.head()
```

```python
rank_gt11 = songs_df["popularity"] > 75
df = songs_df[rank_gt11]
df = df.sort_values(by=["avgcompscore"], axis=0, ascending=False)
df.head(5)
```

```python
#plt.figure(figsize=(10,10))
plt.xlabel("Compound Score")
plt.ylabel("Top Artist(Popularity over 80)")
plt.scatter(df["followers_count"],df["avgcompscore"],color="red")
plt.show()
```

```python
pop1=songs_df['followers_count']>1000000
pop=songs_df[pop1]
pop=pop.sort_values(by=["followers_count"], axis=0, ascending=True)
#pop.head()
#plt.xlim(1000,1000000)
plt.xlabel("Followers Count")
plt.ylabel()
plt.figure(figsize=(10,10))
plt.barh(pop["genres"],pop["followers_count"])
#plt.autoscale(enable=True, axis='both', tight=None)

plt.show()
```

```python
df = songs_df.groupby("year")["avgcompscore"].mean()
plt.figure(figsize=(8,6))
plt.xlabel("Year")
plt.ylabel("Average Compound Score")
plt.title("Sentimental Analysis over the year")
df.plot()
plt.show()
```

```python
plt.scatter(songs_df["avgcompscore"],songs_df["year"])
plt.ylim(1990,2000)
plt.xlim(-1,1)
plt.show()
```

```python
df = songs_df.groupby("genres")["avgcompscore"].mean().reset_index()
df.head()
```

```python
plt.figure(figsize=(10,12))
rank_gt11 = songs_df["popularity"] > 75
df = songs_df[rank_gt11]
df = df.sort_values(by=["popularity"], axis=0, ascending=False)
df.head()
plt.xlabel("Average Compound Score")
plt.ylabel("Genres of Popular Artists")
plt.title("Genres vs Average Compound Score")
plt.barh(df["genres"],df["avgcompscore"])

plt.show()
```
