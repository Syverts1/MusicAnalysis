```python
csv_path = "artist.csv"
df=pd.read_csv(csv_path,encoding='UTF-8')
df.head()
```

```python
k=0

song_data={'id':[],'title':[],'song_count':[],'followers_count':[],'genres':[],'popularity':[]}
```

```python
for i in df['title']:
    k=k+1
    result = sp.search(i,limit=1, offset=0, type='artist', market="US")
    #pprint(result)
    sub=result['artists']['items']
    for follow in sub:
        followerscount = sub[0]["followers"]["total"]
        genre= sub[0]["genres"]
        popular=sub[0]["popularity"]
        title=sub[0]['name']
        if followerscount !=0 or genre !=[""] or popular !=0:
            song_data['followers_count'].append(followerscount)
            song_data['genres'].append(genre)
            song_data['popularity'].append(popular)
            
            #song_data['id']=song_data['id']
            song_data['title'].append(title)
            #song_data['song_count']=df['song_count']
        next
        print(k, i,title)
    next
```

```python
cleaned_csv = zip(song_data['title'],song_data['followers_count'],song_data['genres'],song_data['popularity'])
print(cleaned_csv)

```

```python
output_file = os.path.join("artist_final.csv")
with open(output_file, "w", newline="", encoding="UTF-8") as datafile:
    writer = csv.writer(datafile)
    writer.writerow(["title","followers_count","genres","popularity"])
    writer.writerows(cleaned_csv)
```
